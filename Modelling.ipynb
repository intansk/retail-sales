#### **Penanganan Outliers**

```python
results = []
cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)
display(results_df)
```

# **MODELLING ASSOCIATE**

## **Select Modelling Techniques**

Berdasarkan tujuan dari analisis yang dilakukan, analisis pada dataset ini bertujuan untuk menganalisis pola pembelian yang dilakukan oleh pelanggan sehingga dapat mengetahui tren penjualan. Hal ini bertujuan untuk mengidentifikasi kelompok pelanggan dengan preferensi yang serupa agar strategi pemasaran dan layanan pelanggan dapat disesuaikan secara lebih efektif. Oleh karena itu, teknik permodelan yang digunakan adalah **klastering (asosiatif)** untuk mengelompokkan pelanggan ke dalam segmen-segmen yang berbeda berdasarkan pola pembelian mereka.

### **Algoritma yang Digunakan**

Pada analisis dataset ini, digunakan dua algoritma diantaranya:
1.   **K-Means;** digunakan untuk mengelompokkan data pelanggan ke dalam klaster yang memiliki karakteristik pembelian serupa. Dalam konteks analisis ini, K-Means dapat membantu mengidentifikasi kelompok pelanggan yang memiliki kecenderungan pembelian yang mirip. Misalnya, pelanggan dalam klaster tertentu mungkin cenderung membeli produk-produk yang serupa atau memiliki pola pembelian yang sama. Dengan demikian, dapat membantu dalam mengembangkan strategi pemasaran yang lebih tepat dan menyesuaikan penawaran produk yang disesuaikan dengan preferensi pelanggan.


2.   **Hierarchical Clustering;** algoritma ini memungkinkan untuk membangun struktur hierarki dari klaster sehingga memberikan pemahaman yang lebih dalam tentang bagaimana pelanggan dapat dikelompokkan berdasarkan pola pembeliannya. Algoritma ini dapat membantu dalam menyesuaikan strategi pemasaran yang lebih baik dan menargetkan pelanggan dengan lebih tepat berdasarkan karakteristik pembelian yang spesifik.
